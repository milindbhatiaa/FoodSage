### Recommendation System

```python
from fuzzywuzzy import process
import pandas as pd
```

```python
recipes_label = pd.read_csv("RAW_recipes.csv")
classes = open("classes.txt", 'r').read().splitlines()
len(classes)
```

```python
# Running a matching on label name and dish name

recipes_label['name'] = recipes_label['name'].fillna('').astype(str)
def find_best_match(name):
    if name.strip() == '':
        return '-'  
    match, score = process.extractOne(name, classes)
    return match

# Apply the function to create a new column in the DataFrame
recipes_label['best_match_label'] = recipes_label['name'].apply(find_best_match)
recipes_label.to_csv('Fuzzy_Match_Output_Name_to_Name.csv', index=False)
recipes_label.head()
```

### Tailored Recommendations by applying filters

```python
best_match_label = pd.read_csv('Fuzzy_Match_Output_Name_to_Name.csv')
```

```python
best_match_label.drop(columns=['Unnamed: 0','id','contributor_id','tags','description','submitted'],inplace=True)
```

```python
def convert_to_list(x):
    try:
        return ast.literal_eval(x)
    except (SyntaxError, ValueError):
        return None

# Convert string representation to actual list
best_match_label['nutrition'] = best_match_label['nutrition'].apply(convert_to_list)
```

```python
column_names = ['calories', 'total_fat', 'sugar', 'sodium', 'protein', 'saturated_fat']
best_match_label[['calories', 'total_fat', 'sugar', 'sodium', 'protein', 'saturated_fat', '_']] = pd.DataFrame(best_match_label['nutrition'].to_list(), index=best_match_label.index)

# Drop the original 'nutrition' column and the extra column '_' if needed
best_match_label = best_match_label.drop(['nutrition', '_'], axis=1)
```

```python
def customized_recommendations(df, predicted_output=None, minutes=None, n_steps=None, n_ingredients=None,
                             calories=None, total_fat=None, sugar=None, sodium=None, protein=None, saturated_fat=None):
    # Copy the original DataFrame to avoid modifying the original data
    filtered_df = df.copy()

    # Define a list to store conditions for filtering
    conditions = []

    # Add conditions based on predicted_output
    if predicted_output is not None:
        conditions.append(filtered_df['best_match_label'] == predicted_output)

    # Add conditions based on minutes
    if minutes is not None:
        conditions.append(filtered_df['minutes'] <= minutes)

    # Add conditions based on n_steps
    if n_steps is not None:
        conditions.append(filtered_df['n_steps'] <= n_steps)

    # Add conditions based on n_ingredients
    if n_ingredients is not None:
        conditions.append(filtered_df['n_ingredients'] <= n_ingredients)

    # Add conditions based on calories
    if calories is not None:
        conditions.append(filtered_df['calories'] <= calories)

    # Add conditions based on total fat
    if total_fat is not None:
        conditions.append(filtered_df['total_fat'] <= total_fat)

    # Add conditions based on sugar
    if sugar is not None:
        conditions.append(filtered_df['sugar'] <= sugar)

    # Add conditions based on sodium
    if sodium is not None:
        conditions.append(filtered_df['sodium'] <= sodium)

    # Add conditions based on protein
    if protein is not None:
        conditions.append(filtered_df['protein'] >= protein)

    # Add conditions based on saturated fat
    if saturated_fat is not None:
        conditions.append(filtered_df['saturated_fat'] <= saturated_fat)

    # Apply all conditions using logical AND
    if conditions:
        filtered_df = filtered_df[pd.DataFrame(conditions).all(axis=0)]
    

    return filtered_df
```

```python
recommended_recipes = customized_recommendations(
    best_match_label, predicted_output='pork_chop'
)
```
